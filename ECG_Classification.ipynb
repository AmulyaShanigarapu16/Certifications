```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np

# Load the CSV file into a pandas DataFrame
file_path = "/content/drive/MyDrive/mitbih_train.csv"
df = pd.read_csv(file_path, header=None)

# Separate features (ECG signal) and labels
X = df.iloc[:, :-1].values  # All columns except the last one are the signal
y = df.iloc[:, -1].values   # The last column contains the labels

# Function to plot an ECG beat
def plot_ecg_beat(beat, label, beat_number):
    plt.figure(figsize=(10, 3))
    plt.plot(beat, label=f'Beat {beat_number} (Class {int(label)})')
    plt.title(f'ECG Beat Example -  (Class {int(label)})')
    plt.xlabel('Time Steps')
    plt.ylabel('Amplitude')
    plt.legend()
    plt.show()

# Plot one ECG beat from each class
classes = np.unique(y)  # Get unique class labels

for class_label in classes:
    # Get the index of the first beat from the current class
    class_index = np.where(y == class_label)[0][0]
    # Plot the beat
    plot_ecg_beat(X[class_index], class_label, class_index)
```

```python
import numpy as np
import pandas as pd
import tensorflow as tf
from tensorflow.keras import layers, models
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import OneHotEncoder

# Load the train and test datasets
train_data_path = "/content/drive/MyDrive/mitbih_train.csv"
test_data_path = "/content/drive/MyDrive/augmented_mitbih_test.csv"

# Load the CSV files into pandas DataFrames
train_df = pd.read_csv(train_data_path, header=None)
test_df = pd.read_csv(test_data_path, header=None)

# Separate features and labels from the training and testing datasets
X_train = train_df.iloc[:, :-1].values
y_train = train_df.iloc[:, -1].values

X_test = test_df.iloc[:, :-1].values
y_test = test_df.iloc[:, -1].values

# Reshape the data to match the input shape (samples, time_steps, channels)
# Assuming time_steps = number of features, channels = 1 (for 1D signals)
X_train = X_train.reshape(X_train.shape[0], X_train.shape[1], 1)
X_test = X_test.reshape(X_test.shape[0], X_test.shape[1], 1)

# One-hot encode the labels for classification
onehot_encoder = OneHotEncoder(sparse_output=False, handle_unknown='ignore')  # Added 'handle_unknown' argument
y_train = onehot_encoder.fit_transform(y_train.reshape(-1, 1))
y_test = onehot_encoder.transform(y_test.reshape(-1, 1))


# Residual block definition
def residual_block(x, filters, kernel_size=5):
    shortcut = x
    x = layers.Conv1D(filters, kernel_size, padding='same')(x)
    x = layers.ReLU()(x)
    x = layers.Conv1D(filters, kernel_size, padding='same')(x)
    x = layers.add([x, shortcut])
    x = layers.ReLU()(x)
    return x

# Network architecture
def create_model(input_shape, num_classes):
    inputs = layers.Input(shape=input_shape)

    # Initial Conv layer
    x = layers.Conv1D(32, 5, padding='same')(inputs)

    # Residual blocks (5 blocks in total)
    for _ in range(5):
        x = residual_block(x, 32)
        x = layers.MaxPooling1D(pool_size=5, strides=2)(x)

    # Flatten and Fully connected layers
    x = layers.Flatten()(x)
    x = layers.Dense(32, activation='relu')(x)
    x = layers.Dropout(0.5)(x)
    x = layers.Dense(32, activation='relu')(x)
   # x = layers.Dropout(0.5)(x)

    # Output layer with softmax activation
    outputs = layers.Dense(num_classes, activation='softmax')(x)

    # Define model
    model = models.Model(inputs=inputs, outputs=outputs)

    return model

# Example usage:
input_shape = (X_train.shape[1], 1)  # Assuming each beat has time steps and 1 channel
num_classes = y_train.shape[1]  # Number of output classes

model = create_model(input_shape, num_classes)

# Compile the model with cross entropy loss and Adam optimizer
model.compile(optimizer=tf.keras.optimizers.Adam(learning_rate=0.001, beta_1=0.9, beta_2=0.999, decay=0.00075),
              loss='categorical_crossentropy',
              metrics=['accuracy'])

# Train the model
history = model.fit(X_train, y_train,
                    validation_data=(X_test, y_test),
                    epochs=20,
                    batch_size=128)

# Evaluate the model
test_loss, test_acc = model.evaluate(X_test, y_test, verbose=2)
print(f"Test Accuracy: {test_acc:.4f}")
```

```python
import pandas as pd
import numpy as np

from keras import optimizers, losses, activations, models
from keras.callbacks import ModelCheckpoint, EarlyStopping, ReduceLROnPlateau
from keras.layers import Dense, Input, Dropout, Convolution1D, MaxPool1D, GlobalMaxPool1D
from sklearn.metrics import accuracy_score, f1_score
from sklearn.model_selection import train_test_split

# Load the datasets
df_1 = pd.read_csv("/content/drive/MyDrive/ptbdb_normal.csv", header=None)
df_2 = pd.read_csv("/content/drive/MyDrive/ptbdb_abnormal.csv", header=None)

# Combine datasets and split into train/test sets
df = pd.concat([df_1, df_2])
df_train, df_test = train_test_split(df, test_size=0.2, random_state=1337, stratify=df[187])

# Prepare input features and labels
Y = np.array(df_train[187].values).astype(np.int8)
X = np.array(df_train[list(range(187))].values)[..., np.newaxis]

Y_test = np.array(df_test[187].values).astype(np.int8)
X_test = np.array(df_test[list(range(187))].values)[..., np.newaxis]

# Define the MI classifier model using transfer learning
def get_mi_classifier():
    nclass = 1  # Binary classification: MI or No MI
    inp = Input(shape=(187, 1))

    # Frozen layers (pre-trained feature extractor)
    img_1 = Convolution1D(16, kernel_size=5, activation=activations.relu, padding="valid", trainable=False)(inp)
    img_1 = Convolution1D(16, kernel_size=5, activation=activations.relu, padding="valid", trainable=False)(img_1)
    img_1 = MaxPool1D(pool_size=2)(img_1)
    img_1 = Dropout(rate=0.1)(img_1)
    img_1 = Convolution1D(32, kernel_size=3, activation=activations.relu, padding="valid", trainable=False)(img_1)
    img_1 = Convolution1D(32, kernel_size=3, activation=activations.relu, padding="valid", trainable=False)(img_1)
    img_1 = MaxPool1D(pool_size=2)(img_1)
    img_1 = Dropout(rate=0.1)(img_1)
    img_1 = Convolution1D(32, kernel_size=3, activation=activations.relu, padding="valid", trainable=False)(img_1)
    img_1 = Convolution1D(32, kernel_size=3, activation=activations.relu, padding="valid", trainable=False)(img_1)
    img_1 = MaxPool1D(pool_size=2)(img_1)
    img_1 = Dropout(rate=0.1)(img_1)
    img_1 = Convolution1D(256, kernel_size=3, activation=activations.relu, padding="valid", trainable=False)(img_1)
    img_1 = Convolution1D(256, kernel_size=3, activation=activations.relu, padding="valid", trainable=False)(img_1)
    img_1 = GlobalMaxPool1D()(img_1)
    img_1 = Dropout(rate=0.2)(img_1)

    # Fully connected layers for MI classification
    dense_1 = Dense(64, activation=activations.relu, name="dense_1")(img_1)
    dense_1 = Dense(64, activation=activations.relu, name="dense_2")(dense_1)
    dense_1 = Dense(nclass, activation=activations.sigmoid, name="dense_3_mi")(dense_1)

    model = models.Model(inputs=inp, outputs=dense_1)

    # Compile the model
    opt = optimizers.Adam(0.001)
    model.compile(optimizer=opt, loss=losses.binary_crossentropy, metrics=['acc'])

    # Print model summary
    model.summary()

    return model

# Get the MI classifier model
model = get_mi_classifier()

# Load pre-trained weights from the arrhythmia model
model.load_weights("/content/drive/MyDrive/final_model.h5", by_name=True, skip_mismatch=True)

# Setup callbacks for early stopping and learning rate adjustment
file_path = "mi_classifier_cnn_ptbdb.keras"
checkpoint = ModelCheckpoint(file_path, monitor='val_acc', verbose=1, save_best_only=True, mode='max')
early = EarlyStopping(monitor="val_acc", mode="max", patience=5, verbose=1)
redonplat = ReduceLROnPlateau(monitor="val_acc", mode="max", patience=3, verbose=2)
callbacks_list = [checkpoint, early, redonplat]

# Train the model with transfer learning
model.fit(X, Y, epochs=50, verbose=2, callbacks=callbacks_list, validation_split=0.1)

# Load the best model weights after training
model.load_weights(file_path)

# Make predictions on the test set
pred_test = model.predict(X_test)
pred_test = (pred_test > 0.5).astype(np.int8)

# Evaluate the model with F1 score and accuracy
f1 = f1_score(Y_test, pred_test)
acc = accuracy_score(Y_test, pred_test)

print(f"Test F1 score: {f1}")
print(f"Test accuracy score: {acc}")

```
